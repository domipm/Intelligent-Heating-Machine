# Thermodynamical Model: Drying of Wet Flat Surface with Parallel Airflow

In this first model, which serves as an approximation to the real system, we will treat the drying machine as a one dimensional system, in which we have air entering the drying chamber with a constant velocity, temperature, and humidity level (ideally zero absolute humidity), and with a laminar flow. Inside the chamber is our wet solid, a flat and thin plate, which in the case of stainless steel assume contains all of its moisture in a thin layer at the surface. Assuming an air flow parallel (tangential) to the surface of the solid, a heat transfer sets in between the hot air and the colder solid due to the temperature gradient. This increases the temperature of the solid (and water on its surface) and allows for evaporation. During this process, the air gets slightly colder. With evaporation, water molecules from the moisture surface on the solid diffuse into the air in the chamber, and are effectively carried away by the air flow via convection process. The air absorbs water molecules and therefore increases its humidity level. Finally, exiting the system, air that is slightly more cold and moist exits the chamber, and we can measure both its temperature and humidity via sensors.

We are specially interested in modelling the state of the air coming out of the chamber, that is, its level of humidity and temperature. With these values, we can give a good estimate as to the state of the air inside the chamber and subsequently the state of the solid and its moisture content.

### State Characterization

The main input parameters are the temperature of the air being inserted into the chamber, the humidity of the air (either absolute or relative, depending on which one we can measure), and the velocity of the air or inflow. The *input* state is given by $x_{\text{in}} = ({T_d}^{\text{in}}, {H_A}^{\text{in}} = 0)$. 

The input temperature of the air will be equal to the design temperature $T_d$ chosen on the drying machine. We also want the inflowing air to be dry so that a maximum amount of water can be absorbed in it, and therefore the input air absolute humidity must be as close to $H_A = 0$ as possible. 

The velocity of the air $v_{\text{air}}$ is a parameter which we will consider constant during the whole process.

The output state is defined as the state of the air coming out of the drying chamber, which we will consider a good approximation of the actual air inside. It will be characterized by the temperature and humidity of the expelled air, whose velocity in this case should be approximately the same as the input velocity (although this parameter is not relevant for our analysis). The *output* state is given by $x_{\text{out}} = (T^{\text{out}}, {H_A}^{\text{out}})$.

We will assume that the steel has a uniform temperature which changes instantly, meaning no conduction inside the steel. Also, we will consider that the thin water layer on the steel surface completely covers the surface area initially and has a thickness of $50$ micrometers, with its temperature being equal to that of the steel surface.

## Heat Transfer

The main heat transfer process is convection between the hot air coming into contact with the colder steel slab and water. In this case, we are dealing with forced convection, where the air flow is driven through external forces (a fan). In contrast, natural convection is the movement of the air due to temperature and density gradients due only to the heat transfer. In some cases, mixed convection occurs, where both processes happen simultaneously. However, in this model, let's assume only forced convection occurs.

We will also assume that both the trays and the drying chamber walls are adiabatic, meaning they don't exchange heat with the environment nor the air.

This process is governed by the so-called Newton's cooling equation, which states:
$$
\boxed{ \dot{Q}_{\text{conv}} = h\,A\,(T_s - T_{\infty}) }
$$

where:
$$
\begin{align*}
&\dot{Q}: \text{rate of heat transfer} \\
&h: \text{ \textbf{heat transfer coefficient}, which includes all geometrical and fluid-dynamic behaviour of the particular system} \\
&A: \text{ effective surface area in contact with environment} \\
&T_s: \text{ temperature of the surface} \\
&T_{\infty}: \text{ temperature of the airflow outside the heat transfer boundary layer} \\
\end{align*}
$$

Here, two main difficulties present. First, the effective surface area to be heated. In this particular simplification of a drying machine, it is clear that the upper surface of the flat steel material is considered, however in real drying machines a more thorough analysis must be performes. Second, determining the heat transfer coefficient is usually complicated analytically, and involves performing fluid dynamic simulations (which may be necessary also to simulate heat transfer in real processes).

### Heat Transfer Coefficient

Using fluid mechanics, there exists a method of computing the heat transfer coefficient without the need of complex fluid simulations based on correlation theory, where we study the behaviour of the airflow based on various dimensionless parameters, which then through empirical measurements can be related to the heat transfer coefficient.

In particular, we begin defining **Reynold's number** as the ratio between internal and viscous forces. It can be written as:
$$
\boxed{ \text{Re} = \frac{u L}{\nu} = \frac{\rho u L}{\mu} }
$$

where:
$$
\begin{align*}
&\rho: \text{ density of the fluid} \\
&\nu: \text{ kinematic viscosity of the fluid } \\
&\mu: \text{ dynamic viscosity} \\
&u: \text{ airflow speed} \\
&L: \text{ characteristic length, for example in this case, length of the plate}
\end{align*}
$$

Typically, low Reynolds number indicate laminar flow (where air particles travel parallel) and higher values tend to represent more turbulent flows.

Next, **Prandtl's number** quantifies the ratio of moment diffusivity to thermal diffusivity:
$$
\boxed{ \text{Pr} = \frac{\nu}{\alpha} = \frac{c_p \mu}{k} }
$$

where
$$
\begin{align*}
&\alpha: \text{ thermal diffusivity } \\
&k: \text{ thermal conductivity } \\
&c_p: \text{specific heat}
\end{align*}
$$

Finally, **Nusselt's number** is the ratio of total heat transfer to conductive transfer at the boundary of a fluid. It can be written as:
$$
\begin{align*}
\boxed{ \text{Nu} = \frac{hL}{k} }
\end{align*}
$$

from which we can obtain the convective heat transfer coefficient $h$ using a correlation. For example, the average Nusselt number for laminar flow over a flat plate, from the edge of the plate to a specific downstream distance is:
$$
\overline{\text{Nu}} = \frac{hL}{k} = 0.664\,\text{Re}^{1/2}\,Pr^{1/3} \implies \boxed{ h = \frac{0.664\,k\,\text{Re}^{1/2}\,\text{Pr}^{1/3}}{L} }
$$

Let's calculate the convective heat transfer coefficient for this simple case. Assume inflow air is pure dry air, with velocity $u = 1\.\text{m/s}$. Assuming atmospheric pressure, the density of air at $90\,\text{ÂºC}$ is $\rho = 0.973\,\text{kg/m}^3$. Suppose a characteristic plate length of $L = 0.1\,\text{m}$, and a dynamic viscosity of $\mu = 2.131\cdot10^{-5}\,\text{kg/m/s}$. The Reynolds number is therefore $\text{Re} = 4570$ (This would indicate a turbulent flow!).

Prandtl number for air at this temperature is $\text{Pr} = 0.7132$.

Therefore, the convective heat transfer coefficient is:
$$
h = 0.664\,0.03024\,4570^{1/2}\.0.7132^{1/3} = 1.21144
$$

(However, this correlation is valid for laminar flow, which the Reynolds number tells us its not!).

### Evaporation

Some of the heat transferred to the water surface will be used for evaporation, the breaking of intermolecular forces maintaining the fluid together, and some molecules will become unbound and travel, through convection and concentration diffusion, towards the air stream, which will them carry them outwards. This is characterized with the following rate of latent heat transfer:
$$
\boxed{ \dot{Q}_{\text{vap}} = L_v(T)\,\dot{m}_v}
$$

where:
$$
\begin{align*}
&L_v(T): \text{ latent heat of vaporization, temperature dependent althoug for water we can approximate $L_v \approx 2.26\cdot10^{6}$ [J/kg]} \\
&\dot{m}_v: \text{ rate of mass of water evaporated}
\end{align*}
$$

### Heating of Wet Solid

Another part of the heat transferred into the wet solid will be used to increase its temperature. This is quanized by the specific heat capacity, which relates heat transfer with temperature changes:
$$
\boxed{ \dot{Q}_{\text{temp}} = m_s\,c_s\,\Delta T }
$$

where:
$$
\begin{align*}
&m_s: \text{ mass of the solid} \\
&c_s: \text{ specific heat capacity of the solid object}
\end{align*}
$$

### Energy Balance

The First Law of Thermodynamics is an energy conservation law which states that the internal energy of a system is equal to the work performed on or by the system and the heat transferred into or out of it:
$$
\boxed{ \Delta U = Q - W }
$$

In our case, there is no work performed on or by the system. Since energy must be conserved, the change of internal energy must be zero between the initial and final states, and therefore, the heat transferred into the solid must be equal to that which the solid uses for evaporation and heating. Therefore:
$$
\dot{Q}_{\text{conv}} = \dot{Q}_{\text{temp}} + \dot{Q}_{\text{evap}} \implies h\,A\,\frac{dT(t)}{dt} = L_v(T)\,\dot{m}_v(t) + mc\,\frac{dT(t)}{dt} \implies \boxed{ \frac{dT(t)}{dt}\left( hA - mc \right) = L_v(T)\dot{m}_v(t) }

## Mass Transfer

We will assume that the water content is spread out uniformly as a thin layer over all the surface of the object. Over time due to evaporation the thickness of this water layer will decrease, and we also assume it will do so uniformly, until it dries completely. Through evaporation, water molecules transfer via convection and diffusion to the air and then out of the drying chamber.

The mass of the water present on the surface changes over time in a way proportional to the evaporation rate:
$$
\boxed{ \frac{dm_w}{dt} = h_m\,A\,(X_s - X_{\infty}) }
$$

where:
$$
\begin{align*}
&h_m: \text{ mass transfer coefficient, also depends on geometry similarly to heat transfer coefficient} \\
&A: \text{ area} \\
&X_s: \text{ absolute humidity or humidity ratio at surface} \\
&X_{\infty}: \text{ humidity ratio of air flow}
\end{align*}
$$

In our simulation, the absolute humidity of incoming airfow will be zero (completely dry air), while that of the liquid layer will be saturated. The volume of the thin water layer will be the area of the surface (always considering effective availible area for drying) $A$ multiplied by its thickness $\delta$. At atmospheric pressure, this volume of water will have a total mass of:
$$
\boxed{ m_w(t=0) = A\rho\delta }
$$

and it will steadily decrease as the drying progresses, reaching zero at the end of the drying process. Since it is spread out uniformly, we assume the area $A$ will remain constant.
